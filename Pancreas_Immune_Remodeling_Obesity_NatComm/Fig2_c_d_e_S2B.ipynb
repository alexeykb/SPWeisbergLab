```python
#display(cells_type_ph)
macs_ductal = cells_type_ph_temp.loc[cells_type_ph_temp['Tissue Category']=='Ductal' , ['Exp','BMI','level_3','Phenotype_original']]
macs_exocrine =  cells_type_ph_temp.loc[cells_type_ph_temp['Tissue Category']=='Exocrine', ['Exp','BMI','level_3','Phenotype_original']]
macs_islet =  cells_type_ph_temp.loc[cells_type_ph_temp['Tissue Category']=='Islet', ['Exp','BMI','level_3','Phenotype_original']]

macs_ductal = macs_ductal.loc[(macs_ductal['level_3']=='Mac_CD11c_l') |
                              (macs_ductal['level_3']=='Mac_CD11c_h'),
                              ['Exp','BMI','level_3','Phenotype_original']]
macs_exocrine = macs_exocrine.loc[(macs_exocrine['level_3']=='Mac_CD11c_l') |
                                 (macs_exocrine['level_3']=='Mac_CD11c_h'),
                                 ['Exp','BMI','level_3','Phenotype_original']]
macs_islet = macs_islet.loc[(macs_islet['level_3']=='Mac_CD11c_l') |
                            (macs_islet['level_3']=='Mac_CD11c_h'),
                            ['Exp','BMI','level_3','Phenotype_original']]

macs_ductal = macs_ductal.sort_values(by = 'BMI')
macs_exocrine = macs_exocrine.sort_values(by = 'BMI')
macs_islet = macs_islet.sort_values(by = 'BMI')

#display(macs_ductal)
#display(macs_exocrine)
#display(macs_islet)


macs_ductal = macs_ductal.groupby(['Exp', 'BMI', 'level_3'])['Phenotype_original'].sum()
macs_exocrine = macs_exocrine.groupby(['Exp', 'BMI', 'level_3'])['Phenotype_original'].sum()
macs_islet = macs_islet.groupby(['Exp', 'BMI', 'level_3'])['Phenotype_original'].sum()

#display(macs_ductal)
#display(macs_exocrine)
#display(macs_islet)
#print(macs_ductal.to_string())
#print(macs_exocrine.to_string())
#print(macs_islet.to_string())

macs_ductal = macs_ductal.reset_index()
macs_exocrine = macs_exocrine.reset_index()
macs_islet = macs_islet.reset_index()
print('----------------------->')
print('Ductal: ', len(macs_ductal))
display(macs_ductal)
print('Exocrine: ', len(macs_exocrine))
display(macs_exocrine)
print('Islet: ', len(macs_islet))
display(macs_islet)
print('<-----------------------')

# how depend on the size of DF. Should be used the order of the one with higehr dimension, in order to 
# kepe the values!



macs = pd.merge(macs_ductal, macs_exocrine, how ='left', on = ['Exp', 'BMI','level_3'])
macs = pd.merge(macs, macs_islet, how ='left', on = ['Exp', 'BMI','level_3'])
macs = macs.rename(columns={'Phenotype_original_x': 'CT#_ductal', 
                            'Phenotype_original_y': 'CT#_exocrine',
                            'Phenotype_original': 'CT#_islet'})
#display(macs)
macs = macs.fillna(0)
print(macs.to_string())
```

```python
macs_CD11c_h = macs.loc[macs['level_3'] == 'Mac_CD11c_h']
macs_CD11c_l = macs.loc[macs['level_3'] == 'Mac_CD11c_l']
macs_CD11c_h[['CT#_ductal', 'CT#_exocrine', 'CT#_islet']] = macs_CD11c_h[['CT#_ductal', 'CT#_exocrine', 'CT#_islet']].astype(float)
macs_CD11c_l[['CT#_ductal', 'CT#_exocrine', 'CT#_islet']] = macs_CD11c_l[['CT#_ductal', 'CT#_exocrine', 'CT#_islet']].astype(float)

#display(macs_CD11c_h)
#display(macs_CD11c_l)
print('len macs CD11c+', len(macs_CD11c_h))
print('len macs CD11c-', len(macs_CD11c_l))

#---------------------------------------------------------
#
# merge back and subdivide again to the Null in case there are a donor with only one subtipe (only with macs_CD11c h!)
#
#---------------------------------------------------------
macs_merge = pd.merge(macs_CD11c_h, macs_CD11c_l, how ='right', on = ['Exp', 'BMI'])
print(macs_merge.to_string())



macs_ratio = pd.DataFrame()
macs_ratio['BMI'] = macs_merge['BMI']
macs_ratio['CT#_ductal'] = macs_merge['CT#_ductal_x'].values/(macs_merge['CT#_ductal_x'].values+
                                                              macs_merge['CT#_ductal_y'].values)
macs_ratio['CT#_exocrine'] = macs_merge['CT#_exocrine_x'].values/(macs_merge['CT#_exocrine_x'].values+
                                                                  macs_merge['CT#_exocrine_y'].values)
macs_ratio['CT#_islet'] = macs_merge['CT#_islet_x'].values/(macs_merge['CT#_islet_x'].values+ 
                                                            macs_merge['CT#_islet_y'].values)



macs_ratio['CT#_Comb'] = (macs_merge['CT#_ductal_x'].values + 
                          macs_merge['CT#_exocrine_x'].values +
                          macs_merge['CT#_islet_x'].values)/(macs_merge['CT#_ductal_x'].values+ 
                                                             macs_merge['CT#_ductal_y'].values +
                                                             macs_merge['CT#_exocrine_x'].values+
                                                             macs_merge['CT#_exocrine_y'].values +
                                                             macs_merge['CT#_islet_x'].values+ 
                                                             macs_merge['CT#_islet_y'].values)


#display(macs_ratio)



import seaborn as sns
#macs_ratio = macs_ratio.set_index('BMI')
macs_ratio_heatmap_df = macs_ratio[['BMI','CT#_ductal','CT#_exocrine','CT#_islet', 'CT#_Comb']]
#macs_ratio_heatmap_df['BMI'] = macs_ratio_heatmap_df.index
macs_ratio_heatmap_df = macs_ratio_heatmap_df.reset_index()
#print('-----')
#display(macs_ratio_heatmap_df)
macs_ratio_heatmap_df['BMI'] = macs_ratio_heatmap_df['BMI'].round(2)
macs_ratio_heatmap_df = macs_ratio_heatmap_df.set_index('BMI')
#macs_ratio_heatmap_df = macs_ratio_heatmap_df[['CT#_ductal', 'CT#_exocrine', 'CT#_islet','CT#_Comb' ]]
macs_ratio_heatmap_df = macs_ratio_heatmap_df[['CT#_ductal', 'CT#_exocrine', 'CT#_islet' ]]
macs_ratio_heatmap_df = macs_ratio_heatmap_df.sort_index()
print('------------------')


macs_ratio_heatmap_df = macs_ratio_heatmap_df.rename(columns = {'CT#_ductal':'Ductal', 
                                                          'CT#_exocrine':'Exocrine',
                                                          'CT#_islet':'Islet'})

display(macs_ratio_heatmap_df)

macs_ratio_heatmap_df.to_csv(notebook_path +'/'+ current_condition + '_Num_cells_ratio_subcompartments.csv', header = True)

display(macs_ratio_heatmap_df)




ax = plt.axes()
ax.set_title('# CD11c+ Mac/ # Total Mac')
sns.heatmap(macs_ratio_heatmap_df, cmap='RdYlGn_r', linewidths=0.5, vmin=0, vmax=1,  annot_kws = {'size': 8}, annot=True)
plt.savefig(notebook_path +'/'+ current_condition +'_'+'Number_cells'+'_'+'HEATMAP.pdf'); 

plt.show()






#### -----------------------------------------
#
# Scatterplot of the heatmap
#
#### -----------------------------------------

import scipy.stats as stats


cond = macs_ratio_heatmap_df.columns.values
print(cond)
#print(macs_ratio_heatmap_df.index)
macs_ratio_heatmap_df['BMI'] = macs_ratio_heatmap_df.index


for i in range(len(cond)):
    print(i)
    #print(range(len(cond)))
    #print(macs_ratio_heatmap_df['BMI'])
    #print(macs_ratio_heatmap_df[cond[i]])
    #print(len(macs_ratio_heatmap_df[cond[i]]))
    macs_ratio_heatmap_df_ = macs_ratio_heatmap_df[[cond[i]]]
    #display(macs_ratio_heatmap_df)
    macs_ratio_heatmap_df_ = macs_ratio_heatmap_df_.reset_index()
    #display(macs_ratio_heatmap_df_)
    #print(macs_ratio_heatmap_df_['BMI'])
    print(cond[i])
    #display(macs_ratio_heatmap_df_[cond[i]])
    #(x = macs_ratio_heatmap_df_['BMI'], 
             #     y = macs_ratio_heatmap_df_[cond[i]],
        
    macs_ratio_heatmap_df_temp = macs_ratio_heatmap_df_.replace([np.inf, -np.inf], np.nan).dropna()
    gg=sns.lmplot(x = 'BMI', 
                  y = cond[i],
                  data =  macs_ratio_heatmap_df_temp, 
                  line_kws = {'color' : 'black'},
                  scatter_kws={"s": 10, 'color' : 'black'})
    gg.fig.suptitle(cond[i], fontsize = 16)
    gg.set_axis_labels("BMI", "# CD11c+ Mac/ # Total Mac")
    #print(macs_ratio_heatmap_df_[cond[i]].replace([np.inf, -np.inf], np.nan).dropna())
    
    #display(macs_ratio_heatmap_df_temp)
    r, p = stats.pearsonr(macs_ratio_heatmap_df_temp['BMI'],macs_ratio_heatmap_df_temp[cond[i]])
    print(' p = ', p)
    if p < 0.05:
        fontweight_v='bold'
        fontsize_v = 11
        TEXT = 'r = {:.2f}, p = {:.2g} *'
    else:
        fontweight_v='normal'
        fontsize_v = 10    
        TEXT =' r = {:.2f}, p = {:.2g}'
    plt.annotate(TEXT.format(r, p), 
                 xy=(0.7, 0.9), 
                 xycoords='axes fraction',
                 fontweight = fontweight_v, 
                 fontsize = fontsize_v)
    gg.savefig(notebook_path +'/'+ current_condition +'_'+'number_cells'+'_'+str(cond[i])+'scatter.pdf');
    #plt.annotate(TEXT.format(p), xy=(0.7, 0.85), xycoords='axes fraction')

#plt.xticks(rotation=90)
```
