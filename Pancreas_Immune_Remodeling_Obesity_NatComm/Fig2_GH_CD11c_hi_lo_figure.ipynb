# Figure 2G / 2H - CD11c-high / CD11c-low macrophage neighborhoods vs BMI

This notebook produces Figure 2G (percentage of CD8 T cells within 20 um of CD11c-low macrophages versus BMI) and Figure 2H (percentage of CD8 T cells within 20 um of CD11c-high macrophages versus BMI). Each panel is a linear regression across n = 32 donors, summarized as Pearson r, R-squared = r^2, and p.  CD11c-high / CD11c-low results are shown; th

**Verified regression values.** Figure 2G (Mac_lo_CD11c, CD8): r = 0.018, R^2 = 0.000, p = 0.922, n = 32. Figure 2H (Mac_hi_CD11c, CD8): r = 0.361, R^2 = 0.131, p = 0.042, n = 32.

**Input files.**

- `_CD8_20_lo_CD11cum_NH_BMI_Summary_um.csv` - per-donor % CD8 within 20 um of Mac_lo_CD11c (CD8); source for Figure 2G.
- `_CD8_20_hi_CD11c_um_NH_BMI_Summary_um.csv` - per-donor % CD8 within 20 um of Mac_hi_CD11c (CD8); source for Figure 2H.
- `_NH_summary_20um.csv` - upstream master neighborhood table from which both per-class summaries are derived.

BMI is joined per donor from `input_file.xlsx`. The three CSV files are written by the upstream neighborhood-gating step (`ED_neighborhood_vector_subset_distance_cutoff_combined_norm_NEU_g_stdev_ZERO` in `vectra_lib.py`).

**Output files.**

- Figure 2H: `_20_um_NB__CD8_hi_cd11c_norm_t_cells_.pdf` (`vectra_lib_PCF.py`, savefig line 6756).
- Figure 2G: `_20_LO_cd11c__um_NB__CD8_norm_t_cells_.pdf` (`vectra_lib_PCF.py`, savefig line 6822).

**Key functions.**

- `vectra_lib.py` - `ED_neighborhood_vector_subset_distance_cutoff_combined_norm_NEU_g_stdev_ZERO` (def line 24940): per-donor Box-Cox CD11c high/low macrophage gating (lambda fit via `scipy.stats.boxcox`, lines 25090 and 25121; split at the transformed-zero threshold) and 20 um T-cell-to-macrophage neighborhood computation; writes `_NH_summary_20um.csv` (line 25752).
- `vectra_lib.py` - `macro_ratio_cutoff_combined_norm_NEU_g_stdev_ZERO` (def line 33864): CD11c-high/low macrophage counts per compartment; writes `Macro_ratio.csv` (line 34326).
- `vectra_lib_PCF.py` - `T_cell_NH` (def line 6489) and `T_cell_NH_total` (def line 7158): compute % CD8 within 20 um of the specified macrophage class versus BMI (`scipy.stats.pearsonr`) and render the 2G / 2H regression plots.

## Verification — Cell A (fresh redraw) vs Cell B (original embedded)

**Cell A** below independently re-reads the per-donor CSVs behind the `T_cell_NH` plots and redraws 2G/2H (scatter + OLS fit + 95% CI). These are *my data drawn fresh*, not the original pipeline.

```python
# === CELL A — independent re-draw of panels 2G & 2H from the per-donor CSVs ===
# Verification plot only (NOT the original pipeline): reads the same per-donor CSVs that
# T_cell_NH used, redraws scatter + OLS fit + 95% CI, prints Pearson r, R^2=r^2, p in each title.
import os, pandas as pd, scipy.stats as st, seaborn as sns, matplotlib.pyplot as plt
DENSITY_DIR = "Density"
PANELS = [("2G", "Mac_lo_CD11c (CD11c-low)",  "_CD8_20_lo_CD11cum_NH_BMI_Summary_um.csv"),
          ("2H", "Mac_hi_CD11c (CD11c-high)", "_CD8_20_hi_CD11c_um_NH_BMI_Summary_um.csv")]
for panel, sub, fn in PANELS:
    d = pd.read_csv(os.path.join(DENSITY_DIR, fn))[["BMI", "T_nh_perc"]].dropna()
    lr = st.linregress(d["BMI"], d["T_nh_perc"])
    fig, ax = plt.subplots(figsize=(4.6, 4.2))
    sns.regplot(data=d, x="BMI", y="T_nh_perc", ci=95, ax=ax,
                scatter_kws=dict(s=42, alpha=.8, color="#1f6fb4"), line_kws=dict(color="#c0392b"))
    ax.set_xlabel("BMI"); ax.set_ylabel("% CD8 within 20 um of " + sub)
    ax.set_title("Fig %s - %s\nr=%.3f   R2=r2=%.3f   p=%.3f   n=%d" % (panel, sub, lr.rvalue, lr.rvalue**2, lr.pvalue, len(d)))
    plt.tight_layout(); plt.show()
    print("%s: slope=%.4f  intercept=%.4f  r=%.4f  R2=%.4f  p=%.4f  n=%d" % (panel, lr.slope, lr.intercept, lr.rvalue, lr.rvalue**2, lr.pvalue, len(d)))
```
